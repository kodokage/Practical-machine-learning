```python
import pandas as pd
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
covid_data = pd.read_csv('covid_19_data.csv')
```

# Date with highest occurrence

```python
covid_data
```

# maximum value

```python
covid_data.max()
```

```python
covid_data.max().ObservationDate
```

# Rate of Recovery

```python
plt.plot(covid_data.ObservationDate, covid_data.Recovered)
plt.title("Rate of recovery")
plt.xlabel("Observaion Date")
plt.ylabel("Recovered")
```

```python
plt.stem(covid_data.ObservationDate, covid_data.Recovered)

plt.show()
```

```python
covid_data.Deaths.max()
```

```python
country = np.asarray(covid_data.max())
```

# Rate of Death

```python
plt.plot(covid_data.ObservationDate, covid_data.Deaths)
plt.title("Rate of Death")
plt.xlabel("Observaion Date")
plt.ylabel("Deaths")
```
